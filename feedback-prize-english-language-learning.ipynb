```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Memuat data
train_data = pd.read_csv('/kaggle/input/feedback-prize-english-language-learning/train.csv')

print(train_data.head())
print(train_data.info())
print(train_data.describe())

# Visualisasi distribusi nilai pada target(cohesion, syntax, vocabulary, phraseology, grammar, conventions)
plt.figure(figsize=(12, 8))
sns.histplot(train_data['cohesion'], bins=20, kde=True, color='blue', label='cohesion')
sns.histplot(train_data['syntax'], bins=20, kde=True, color='green', label='syntax')
           
plt.title('Distribution of Analytic Measures')
plt.xlabel('score')
plt.ylabel('frequency')
plt.legend()
plt.show()


```
